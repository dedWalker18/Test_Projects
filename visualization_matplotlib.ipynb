### Styling Lines and Markers

The `plt.plot` function supports many arguments for styling lines and markers:

- `color` or `c`: Set the color of the line ([supported colors](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
- `linestyle` or `ls`: Choose between a solid or dashed line
- `linewidth` or `lw`: Set the width of a line
- `markersize` or `ms`: Set the size of markers
- `markeredgecolor` or `mec`: Set the edge color for markers
- `markeredgewidth` or `mew`: Set the edge width for markers
- `markerfacecolor` or `mfc`: Set the fill color for markers
- `alpha`: Opacity of the plot


Check out the documentation for `plt.plot` to learn more: [https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) .

```python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x', ls='--')
plt.xlabel('Years')
plt.ylabel('Yield')

plt.title('Crop Yield in Kanto')
plt.legend(['Apples', 'Oranges']);
```

```python
setosa_df = flower_df[flower_df.species == 'setosa']
versicolor_df = flower_df[flower_df.species == 'versicolor']
virginica_df = flower_df[flower_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width,color='b', alpha=0.4, bins=np.arange(2,5,0.5))
plt.hist(versicolor_df.sepal_width, color='r', alpha=0.4, bins=np.arange(2,5,0.5));
```

```python
plt.title("Distribution of Sepal Width")
plt.hist([setosa_df.sepal_width, virginica_df.sepal_width, versicolor_df.sepal_width], color=['r','g','b'] ,
         alpha=0.5, 
         bins=np.arange(2,5,0.4),
         stacked=True)
plt.legend(['Setosa', 'Virginica', 'Versicolor'])
plt.xlabel('Sepal Width')
plt.ylabel('Count');
```

# with matplotlib you can call axes directly but with sns you need to specify it using ax attribute as shown below

```python
fig, axes = plt.subplots(2, 3, figsize=(16,9))

axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years, oranges, 'o--r')
axes[0,0].set_title('Yield')
axes[0,0].legend(['Apples', 'Oranges'])
axes[0,0].set_xlabel('Years')
axes[0,0].set_ylabel('Yield')

axes[0,1].set_title('Sepal Width/Length')
sns.scatterplot(flower_df.sepal_length,
                flower_df.sepal_width,
                hue = flower_df.species,
                s = 100,
                ax = axes[0,1],
                legend=False)
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.sepal_width, virginica_df.sepal_width, versicolor_df.sepal_width],
               bins = np.arange(2, 5, 0.5),
               stacked = True)
axes[0,2].legend(['Setosa', 'Virginica', 'Versicolor'])

axes[1,0].set_title('Restuarant Bills')
sns.barplot('day', 'total_bill', hue = 'sex', data = tips_df, ax = axes[1,0]);

axes[1,1].set_title('Flights traffic')
sns.heatmap(flight_matrix_df, cmap='Blues', ax=axes[1,1]);

axes[1,2].set_title('Data Science Meme')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].axis('off')
axes[1,2].set_xticks([])
axes[1,2].set_yticks([]);


plt.tight_layout(pad=2);
```
